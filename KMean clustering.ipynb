```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cf_df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/Cust_Segmentation.csv")
cf_df
```

```python
df=cf_df.drop('Address',axis=1)
df
```

```python
x=df.values[:,1:]
x=np.nan_to_num(x)#replace a + or- no inspite of nan_values
cluster_data=StandardScaler().fit_transform(x)
cluster_data[0:29]
```

```python
k = 3  
k_means = KMeans(init="k-means++", n_clusters=k, n_init=13)
k_means.fit(x)
```

```python
labels=k_means.labels_
print(labels[0:13])
```

```python
df["clus_km"]=labels
df.head(3)
```

```python
df.groupby('clus_km').mean()
```

```python
area=np.pi*(x[:,1])**2
plt.scatter(x[:,0],x[:,3],s=area,c=labels.astype(np.float),alpha=0.5)
plt.xlabel("Age",fontsize=16)
plt.ylabel("Income",fontsize=16)
plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
plt.clf()
ax=Axes3D(fig,rect=[0,0,.95,1],elev=28,azim=134)
plt.cla()

ax.set_xlabel('Education')
ax.set_ylabel("Age")
ax.set_zlabel("Income")
ax.scatter(x[:,1],x[:,2],x[:,3],c=labels.astype(np.float))
plt.show()
```
